 # Project 3: Customer Segmentation (using K-Means Clustering)

```python
import numpy as np 
import matplotlib.pyplot as plt 

from sklearn.cluster import KMeans

# Create a small "income vs spending score" dataset:
np.random.seed(42)

#pretend: [Annual_Income,Spdending_Score]
customer_data = np.vstack([
    np.random.normal(loc=[30,30],scale=[5,5],size=(50,2)),   # low income, low spend
    np.random.normal(loc=[60,60],scale=[5,5],size=(50,2)),   # mid income, mid spend
    np.random.normal(loc=[90,40],scale=[5,5],size=(50,2)),]) # High income , high spend

print("shape:", customer_data.shape)
```

```python
# Step 2: Visualize raw customer

plt.figure(figsize=(6,4))
plt.scatter(customer_data[:,0], customer_data[:,1])
plt.xlabel("Annual income")
plt.ylabel("Spending Score")
plt.title("Customers (Before clustering)")
plt.grid(True, linestyle="--",alpha=0.4)
plt.show()
```

```python
# Step 3: Run K-Means with k=3 
k = 3
Kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
Kmeans.fit(customer_data)

labels = Kmeans.labels_
centers = Kmeans.cluster_centers_
```

```python
# Step 4: Visualize Clusters + centroids
plt.figure(figsize=(6, 4))

for cluster_id in range(k):
    cluster_data = customer_data[labels == cluster_id]
    plt.scatter(
        cluster_data[:, 0], cluster_data[:, 1],
    label=f"Cluster {cluster_id}"
    )
plt.scatter(
    centers[:,0],
    centers[:,1], 
    edgecolors="black",
    marker='X',
    s=200, 
    linewidths=2,
    label="centroids"
)

plt.xlabel("Annual Income (scaled)")
plt.ylabel("Spending Score")
plt.title("Customers Segments using K-Means")
plt.legend()
plt.grid(True, linestyle="-", alpha=0.4)
plt.show()
```
